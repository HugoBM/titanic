## ANN
### Importing the Keras Libraries

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization
from keras.wrappers.scikit_learn import KerasClassifier
from keras.optimizers import Adam
from keras.callbacks import LearningRateScheduler, TerminateOnNaN
```

### Constructing the Architecture of the ANN

```python
# Initializing the ANN
classifier_ann = Sequential()
# Adding the input layer and the first hidden layer w/ Dropout (to avoid overfitting!)
# Number of nodes in hidden layer could be the average of input + output layers
classifier_ann.add(Dense(units=65, kernel_initializer='uniform', activation='relu', input_dim=X_train.shape[1]))
classifier_ann.add(Dropout(rate=0.25))
# Adding the second hidden layer w/ Dropout
classifier_ann.add(Dense(units=65, kernel_initializer='uniform', activation='relu'))
classifier_ann.add(Dropout(rate=0.25))
# Adding the third hidden layer w/ Dropout
classifier_ann.add(Dense(units=65, kernel_initializer='uniform', activation='relu'))
classifier_ann.add(Dropout(rate=0.25))
# Adding the fourth hidden layer w/ Dropout
classifier_ann.add(Dense(units=65, kernel_initializer='uniform', activation='relu'))
classifier_ann.add(Dropout(rate=0.25))
# Adding the output layer
classifier_ann.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
# Compiling the ANN (same logistic regression: binary_crossentropy or categorical_crossentropy)
classifier_ann.compile(optimizer=Adam(lr=1e-4), loss='binary_crossentropy', metrics=['accuracy'])
# Create learning rate scheduler to decrease learning rate over time
annealer = LearningRateScheduler(lambda e: 1e-3 * 0.85 ** e)
```

### Fitting the ANN to the Training set and making predictions on Test set

```python
hist = classifier_ann.fit(X, y, validation_split=.1, 
                          batch_size=32, epochs=32, callbacks=[annealer, TerminateOnNaN()], verbose=0)
predictions_ann = (classifier_ann.predict(X_test) > 0.5).astype(int).reshape(-1,)
```

```python
def batch_ensemble(classifier_list, X_train, X_test, y_train, num_batches=20):
    """
    Takes a list of classifiers, and a Training set and creates random batches
    of features and observations on which each classifier is trained. Then it fits the 
    classifier to the entire Test set.
    Returns a vector of predictions where each value is the mode of the predictions for
    each row in the Test set.
    """
    nrow_batch = int(np.sqrt(num_batches))
    ncol_batch = int(np.sqrt(num_batches))
    row_batch_size = 300
    col_batch_size = 20

    pred_accumulator = np.zeros((X_test.shape[0], 1))
    j=1
    
    for col_batch in range(ncol_batch):
        col_idx = np.random.choice(a=X_train.shape[1], size=col_batch_size, replace=False)
        print('Batch number {}/{}... done.'.format(j, num_batches))
        j += int(np.sqrt(num_batches)) + 1
        for row_batch in range(nrow_batch):
            row_idx = np.random.choice(a=X_train.shape[0], size=row_batch_size, replace=False)
            for clf in classifier_list:
                clf.fit(X_train[row_idx.reshape(-1, 1), col_idx], y_train[row_idx])
                pred_accumulator = np.append(pred_accumulator, clf.predict_proba(X_test[:, col_idx])[:, 1].reshape(-1, 1), axis=1)
    print('Batch number {}/{}... done.'.format(num_batches, num_batches))
    
    return np.mean(pred_accumulator[:, 1:], axis=1)
```
